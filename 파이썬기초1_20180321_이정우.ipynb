#### 문자열 슬라이싱

```python
a[12:17] # 12 <= a < 17, 'short'
```

```python
a[-6]+a[-5]+a[-4]+a[-3]+a[-2]+a[-1]
```

```python
a[-6:] # 처음부터 혹은 끝까지는 콜론(:)
```

```python
a[:] # 처음부터 끝까지 모두
```

```python
# 연월일과 날씨정보가 한 번에 입력되었다.
a = "20180321Sunny"
date = a[:8] # 날짜만 슬라이싱
weather = a[8:] # 날씨만 슬라이싱
date
```

#### 문자열 포매팅
- 문자열 내의 특정한 값을 바꿔야 할 경우가 있는데 이 때 사용한다.
- 쉽게 말해 문자열 내에 어떤 변하는 값을 삽입하는 방법이다.

```python
# 숫자 바로 대입
"I eat {0} apples" .format(3)
```

```python
# 문자열 바로 대입
"I eat {0} apples" .format("five")
```

```python
# 숫자 값을 가진 변수로 대입하기
number = 3
"I eat {0} apples" .format(number)
```

```python
# 2개 이상의 값 넣기
number = 10
day = "there"
"I ate {0} apples. So I was sick for {1} days." .format(number, day)
```

```python
# 이름으로 넣기
"I ate {number} apples. So I was sick for {day} days." .format(number=10, day=3)
```

```python
# 인덱스와 이름 혼용하기
"I ate {0} apples. So I was sick for {day} days." .format(10, day=3)
```

```python
# 정렬과 공백 채우기
# 왼쪽 정렬
"{0:<10}" .format("hi")
```

```python
# 오른쪽 정렬
"{0:>10}" .format("hi")
```

```python
# 가운데 정렬
"{0:^10}" .format("hi")
```

```python
# 공백을 =으로 채우고 가운데 정렬
"{0:=^10}" .format("hi")
```

```python
# 공백을 !로 채우고 왼쪽 정렬
"{0:!<10}" .format("hi")
```

```python
# 소수점 표현하기
y = 3.42134234
"{0:0.4f}" .format(y) 
```

```python
# 10칸에 소수점 4째자리까지만
"{0:10.4f}" .format(y)
```

```python
# 10칸에 왼쪽정렬하고 소수점 4째자리까지ㅣ만
"{0:<10.4f}" .format(y)
```

```python
# { 또는 } 표현하기
"{{and}}" .format()
```

```python
"{{and}}" .format(3)
```

#### 문자열 관련 함수
- 문자열 자료형은 자체적으로 가지고 있는 함수들이 있다.
- 이를 다른 말로 문자열 내장함수라고 한다. 이 내장함수를 사용하려면 변수 이름 뒤에 '.'을 붙이고 함수 이름을 써주면 된다.

##### 문자 개수 세기(count)
- 문자열 중 문자 b의 개수를 반환한다.

```python
a = "hobby"
a.count('b')
```

##### 위치 알려주기1(find)
- 문자열 중 문자 b가 처음으로 나온 위치를 반환한다.
- 만약 찾는 문자나 문자열이 존재하지 않는다면 -1을 반환한다.

```python
a = "Python is best choice."
a.find('b')
```

```python
a.find('k')
```

##### 위치 알려주기2(index)
- 문자열 중 문자 t가 맨 처음으로 나온 위치를 반환한다.
- 만약 찾는 문자나 문자열이 존재하지 않는다면 오류를 발생시킨다.
- 앞의 find 함수와 다른 점은 문자열 안에 존재 하지 않는 문자를 찾으면 오류가 발생한다는 점이다.

```python
a.index('t')
```

```python
a.index('k')
```

##### 문자열 삽입(join)
- abcd라는 문자열의 각각의 문자 사이에 변수 a의 값인 ','를 삽입한다.

```python
a = ","
a.join('abcd')
```

##### 소문자를 대문자로 바꾸기(upper)
- upper()함수는 소문자를 대문자로 바꾸어 준다. 
- 만약 문자열이 대문자라면 아무런 변화도 일어나지 않을 것이다.

```python
a = "hi"
a.upper()
```

##### 대문자를 소문자로 바꾸기(lower)
- lower()함수는 대문자를 소문자로 바꾸어준다.

```python
a = "HI                                                                                                                                      "
a.lower()
```

##### 왼쪽 공백 지우기(lstrip)
- 문자열 중 가장 왼쪽에 있는 한 칸 이상의 연속된 공백을 모두 지운다.

```python
a = " hi "
a.lstrip()
```

##### 오른쪽 공백 지우기(rstrip)
- 문자열 중 가장 오른쪽에 있는 한 칸 이상의 연속된 공백들을 모두 지운다.

```python
a.rstrip()
```

##### 양쪽 공백 지우기(strip)
- 문자열 양쪽에 있는 한 칸 이상의 연속된 공백들을 모두 지운다.

```python
a.strip()
```

##### 문자열 바꾸기(replace)
- replace(바뀌게 될 문자열, 바꿀 문자열)처럼 사용해서 문자열 내의 특정한 값을 다른 값으로 치환해 준다.

```python
a = "Life is too short"
a.replace("Life", "Your leg")
```

##### 문자열 나누기(split)
- a.split() 처럼 괄호 안에 아무런 값도 넣어 주지 않으면 공백(스페이스, 탭, 엔터 등)을 기준으로 문자열을 나누어 준다.
- a.split(':')처럼 괄호 안에 특정한 값이 있을 경우에는 괄호 안의 값을 구분자로 해서 문자열을 나누어 준다. 이렇게 나눈 값은 리스트에 하나씩 들어간다.

```python
a = "Life is too short"
a.split()
```

```python
a = "a:b:c:d"
a.split()
```

```python
a.split(':')
```
