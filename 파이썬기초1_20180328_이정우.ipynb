#### 문자열 포매팅
- 문자열 내의 특정한 값을 바꿔야 할 경우가 있는데 이 때 사용한다.
- 쉽게 말해 문자열 내에 어떤 변하는 값을 삽입하는 방법이다.

```python
# 숫자 바로 대입
"I eat {0} apples" .format(3)
```

```python
# 문자열 바로 대입
"I eat {0} apples" .format("five")
```

```python
# 숫자 값을 가진 변수로 대입하기
number = 3
"I eat {0} apples" .format(number)
```

```python
# 2개 이상의 값 넣기
number = 10
day = "there"
"I ate {0} apples. So I was sick for {1} days." .format(number, day)
```

```python
# 이름으로 넣기
"I ate {number} apples. So I was sick for {day} days." .format(number=10, day=3)
```

```python
# 인덱스와 이름 혼용하기
"I ate {0} apples. So I was sick for {day} days." .format(10, day=3)
```

```python
# 정렬과 공백 채우기
# 왼쪽 정렬
"{0:<10}" .format("hi")
```

```python
# 오른쪽 정렬
"{0:>10}" .format("hi")
```

```python
# 가운데 정렬
"{0:^10}" .format("hi")
```

```python
# 공백을 =으로 채우고 가운데 정렬
"{0:=^10}" .format("hi")
```

```python
# 공백을 !로 채우고 왼쪽 정렬
"{0:!<10}" .format("hi")
```

```python
# 소수점 표현하기
y = 3.42134234
"{0:0.4f}" .format(y) 
```

```python
# 10칸에 소수점 4째자리까지만
"{0:10.4f}" .format(y)
```

```python
# 10칸에 왼쪽정렬하고 소수점 4째자리까지ㅣ만
"{0:<10.4f}" .format(y)
```

```python
# { 또는 } 표현하기
"{{and}}" .format()
```

```python
"{{and}}" .format(3)
```

## 튜플 자료형
- 튜플(tuple)은 몇 가지 점을 제외하곤 리스트와 거의 비슷하며 리스트와 다른 점은 다음과 같다.
1. 리스트는 [과 ]으로 둘러싸지만 튜플은 (과 )으로 둘러싼다.

## 딕셔너리 자료형
- 대응관계를 나타낼 수 있는 자료형이다.
- 연관 배열(Associateive array) 또는 해시(Hash)라고 한다.
- 딕셔너리는 리스트나 뉴플처럼 순차적으로 해당 요소값을 구하지 않고 Key를 통해 Value를 얻는다. 

#### 딕셔너리는 어떻게 만들까?
- 기본적인 딕셔너리형은 Key와 Value의 쌍 여러개가 {과 }로 둘러싸여 있다. 
- {Key1:Value1, Key2:Value2, Key3,Value3 ...}
- 각각의 요소는 Key:Value 형태로 이루어져 있고 쉽표(,)로 구분되어 있다.

```python
dic = {'name':'pey', 'phone':'01012345678', 'birth':'1118'}
```

```python
# 숫자형 대 문자열
a = {1: 'hi'}
```

```python
# 문자열 대 리스트
b = {'a':[1,2,3]}
```

##### 딕셔너리에서 Key 사용해 Value 얻기

```python
a = {1:'a', 2:'b'}
a[1]
```

- 따라서 a[1]은 딕셔너리 {1:'a',2:'b'}에서 Key가 1인 것의 Value인 'a'를 돌려주게 된다. a[2] 역시 마찬가지이다.

#### 메모리에 생성된 변수 없애기
- a=3을 입력하면 3이라는 정수형 객체가 메모리에 생성되는데, 이 값을 없앨 수도 있다.
- 3이라는 객체를 가리키는 변수의 수를 reference count라고 했는데, 이 reference count가 0이 되는 순간 3이라는 객체는 자동으로 사라진다.
- 즉 3이라는 객체를 가리키고 있는 것이 하나도 없을 때 3이라는 객체는 메모리에서 사라지게 되는 것이다.
- 이것을 가비지 콜렉션(garbage collection)이라고 한다.

```python
a = 3
b = 3
del(a)
del(b)
```
