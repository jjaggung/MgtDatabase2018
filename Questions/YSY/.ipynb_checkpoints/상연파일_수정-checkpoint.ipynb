```python
import pandas as pd
import numpy as np
```

# 데이터 준비

```python
number = pd.read_csv("number.csv", encoding="utf-8", thousands = ',')
number.head()
```

## 추가 부분
### 오류 발생 원인 
##### 합치려고 하는 Obesity는 '구별' 이고 number는 '구분'으로 되어 있어서 오류가 남. 같은 이름이어야 함.
##### 따라서 '구별'로 통일하거나 '구분으로 통일해야 함.

```python
number.rename(columns={number.columns[0] : '구별'}, inplace=True) # 구분을 구별로 바꿈
number.head()
```

### number는 '합계' , Obesity와 health_total은 '서울시'로 되어 있어서 통일

```python
number.iloc[0,0] = '서울시'
number.head()
```

```python
number['1인가구비율'] = number['1인가구'] / number['계'] * 100
```

```python
del number['2인가구']
del number['3인가구']
del number['4인가구']
del number['5인 가구']
del number['6인가구']
del number['7인이상 가구']

number.head()
```

```python
health_total.rename(columns={health_total.columns[0] : '구별'}, inplace=True)
health_total.head()
```

```python
Obesity = pd.read_csv("Obesity.csv", encoding="utf-8")
Obesity.head()
```

## 데이터를 모두 준비해놓고 merge한 후 set_index를 하는 것이 좋습니다.

```python
data_result = pd.merge(number, Obesity, on="구별")
data_result.tail()
```

```python
# 그래프 그리기를 위한 set_index
number.set_index("구별", inplace=True)
Obesity.set_index("구별", inplace=True)
health_total.set_index("구별", inplace=True)
data_result.set_index("구별", inplace=True)
```

# 그래프 그리기

```python
# 그래프를 그리는 라이브러리
import matplotlib.pyplot as plt
# 그래프의 결과를 출력 세션에 나타나게 하는 설정임.
%matplotlib inline 
```

```python
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
```

# 1인가구 비율 그래프

```python
plt.figure()
plt.title('1인가구비율')
number['1인가구비율'].plot(kind='barh', grid=True, figsize=(10,10)) # 수평 막대 그래프를, 그리드를 사용하여, 그림 크기를 10*10으로 지정함.
plt.show()
```

```python
# sort_values 사용
number['1인가구비율'].sort_values().plot(kind='barh', 
                                     grid=True, figsize=(10,10))
plt.show()
```

```python
import folium
```

```python
# 한국 지도 형태 가져오기
import json
geo_path = 'skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

# 1인가구비율 지도

```python
# 1인가구비율로 그리기
map = folium.Map(location=[37.5502, 126.982], zoom_start=10, 
                 tiles='Stamen Toner')

map.choropleth(geo_data = geo_str,
               data = number['1인가구비율'],
               columns = [number.index, number['1인가구비율']],
               fill_color = 'PuRd', #PuRd, YlGnBu
               key_on = 'feature.id')
map
```
