```python
y_fin=pd.read_excel('Seoul_financial.xlsx')
y_fin.head(6)
```

```python
y_fin.set_index(keys="연도", inplace=True)
```

```python
col = []
for i in y_fin.columns:
    col.append(i.split()[0][:-1]+"."+i.split()[1][:-1])
col[:5]
y_fin.columns = col
```

```python
연도선택 = [2013, 2014, 2015, 2016, 2017] 

xs0 = y_fin.columns # 연도
xs0 = pd.to_datetime(xs0)## 가로를 연도로 보기 위해
plt.figure(figsize=(15,15))

for i in range(len(연도선택)):
    plt.plot(xs0, list(y_fin.loc[연도선택[i],:]), marker= '.', label=연도선택[i]) # 녹색 실선


plt.legend(bbox_to_anchor=(1, 1), loc=0,borderaxespad=0.) # 범례 그리기, loc=9 ; top center를 의미
plt.ylabel('당기순손익')
plt.xlabel('연도')
plt.title('당기순손익의 변화')
plt.show()
```
